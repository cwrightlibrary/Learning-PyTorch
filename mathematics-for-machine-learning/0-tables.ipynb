# Tables

## Symbols

| Symbol | Typcal meaning |
|--------|----------------|
| $a,b,c,\alpha,\beta,\gamma$ | Scalars are lowercase |
| $\mathbf{x,y,z}$ | Vectors are bold lowercase |
| $\mathbf{A,B,C}$ | Matrices are bold uppercase |
| $\mathbf{x}^T,\mathbf{A}^T$ | Transpose of a vector or matrix |
| $\mathbf{A}^{-1}$ | Inverse of a matrix |
| $\langle \mathbf{x,y}\rangle$ | Inner dot product of $\mathbf{x}$ and $\mathbf{y}$ |
| $\mathbf{x}\cdot \mathbf{y}$ | Dot product of $\mathbf{x}$ and $\mathbf{y}$ |
| $\mathbf{B}=(b_1,b_2,b_3)$ | (Ordered) tuple |
| $\mathbf{B}=[b_1,b_2,b_3]$ | Matrix of column vectors stacked horizontally |
| $\mathbf{B}=\{b_1,b_2,b_3\}$ | Set of vectors (unordered) |
| $\mathbb{Z,N}$ | Integers and natural numbers, respectively |
| $\mathbb{R,C}$ | Real and complex numbers, respectively |
| $\mathbb{R}^n$ | $n$-dimensional vector space of real numbers |
| $\forall x$ | Universal quantifier: for all $x$ |
| $\exists x$ | Existential quantifier: there exists $x$ |
| $a:=b$ | $a$ is defined as $b$ |
| $a =: b$ | $b$ is defined as $a$ |
| $a \propto b$ | $a$ is proportional to $b$, i.e., $a=\text{constant}\cdot b$ |
| $g \circ f$ | Function composition: "$g$ after $f$" |
| $\Leftrightarrow$ | If and only if |
| $\Rightarrow$ | Implies | 
| $\mathcal{A,C}$ | Sets |
| $a \in \mathcal{A}$ | $a$ is an element of set $\mathcal{A}$ |
| $\emptyset$ | Empty set |
| $\mathcal{A}\setminus \mathcal{B}$ | $\mathcal{A}$ without $\mathcal{B}$ |
| $D$ | Number of dimensions; indexed by $d=1,...,D$ |
| $N$ | Number of data points; indexed by $n=1,...,N$ |
| $\mathbf{I}_m$ | Identity matrix of size $m\times m$ |
| $\mathbf{0}_{m,n}$ | Matrix of zeros of size $m\times n$ |
| $\mathbf{1}_{m,n}$ | Matrix of ones of size $m\times n$ |
| $\mathbf{e}_i$ | Standard/canonical vector (where $i$ is the component that is $1$) |
| $\dim$ | Dimensionality of vector space |
| $\text{rk}(\mathbf{A})$ | Rank of matrix $\mathbf{A}$ |
| $\text{Im}(\boldsymbol{\Phi})$ | Image of linear mapping $\boldsymbol{\Phi}$ |
| $\ker(\boldsymbol{\Psi})$ | Kernel (null space) of a linear mapping $\boldsymbol{\Psi}$ |
| $\text{span}[b_i]$ | Span (generating set) of $b_i$ |
| $\text{tr}(\mathbf{A})$ | Trace of $\mathbf{A}$ |
| $\det(\mathbf{A})$ | Determinant of $\mathbf{A}$ |
| $\lvert \cdot \rvert$ | Absolute value or determinant (depending on context) |
| $\lvert \lvert \cdot \rvert \rvert$ | Norm; Euclidean, unless specified |
| $\lambda$ | Eigenvalue or Lagrange multiplier |
| $E_\lambda$ | Eigenspace corresponding to eigenvalue $\lambda$ |
| $\mathbf{x}\perp \mathbf{y}$ | Vectors $\mathbf{x}$ and $\mathbf{y}$ are orthogonal |
| $V$ | Vector space |
| $V^\perp$ | Orthogonal complement of vector space $V$ |
| $\sum_{n=1}^N x_n$ | Sum of the $x_n:x_1+...+x_n$ |
| $\prod_{n=1}^N x_n$ | Product of the $x_n:x_1+...+x_n$ |
| $\boldsymbol{\theta}$ | Parameter vector |
| $\frac{\partial f}{\partial x}$ | Partial derivative of $f$ with respect to $x$ |
| $\frac{df}{dx}$ | Total derivative of $f$ with respect to $x$ |
| $\nabla$ | Gradient |
| $f_*=min_x f(x)$ | The smallest function value of $f$ |
| $x_* \in \arg\min_x f(x)$ | The value of $x_*$ that minimizes $f$ (note: $\arg\min$ returns a set of values) |
| $\mathcal{L}$ | Lagrangian |
| $\ell$ | Negative log-likelihood |
| $\binom{n}{k}$ | Binomial coefficient, $n$ choose $k$ |
| $\text{Var}_X[\mathbf{x}]$ | Variance of $\mathbf{x}$ with respect to the random variable $X$ |
| $\mathbb{E}_X[\mathbf{x}]$ | Expectation of $\mathbf{x}$ with respect to the random variable $X$ |
| $\text{Cov}_{X,Y}[\mathbf{x,y}]$ | Covariance between $\mathbf{x}$ and $\mathbf{y}$ |
| $X\perp Y \mid Z$ | $X$ is conditionally independent of $Y$ given $Z$ |
| $X\sim p$ | Random variable $X$ is distributed according to $p$ |
| $\mathcal{N}(\boldsymbol{\mu}, \boldsymbol{\Sigma})$ | Gaussian distribution with mean $\boldsymbol{\mu}$ and covariance $\boldsymbol{\Sigma}$ |
| $\text{Ber}(\mu)$ | Bernoulli distribution with parameter $\mu$ |
| $\text{Bin}(N,\mu)$ | Binomial distribution with parameters $N$, $\mu$ |
| $\text{Beta}(\alpha,\beta)$ | Beta distribution with parameters $\alpha$, $\beta$ |
