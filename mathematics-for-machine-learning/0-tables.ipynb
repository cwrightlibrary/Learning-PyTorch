## Table of Symbols

$$
\begin{array}{|l|l|}
\hline
\textbf{Symbol} & \textbf{Typical meaning} \\
\hline
a, b, c, \alpha, \beta, \gamma & \text{Scalars are lowercase} \\
\mathbf{x}, \mathbf{y}, \mathbf{z} & \text{Vectors are bold lowercase} \\
\mathbf{A}, \mathbf{B}, \mathbf{C} & \text{Matrices are bold uppercase} \\
\mathbf{x}^T, \mathbf{A}^T & \text{Transpose of a vector or matrix} \\
\mathbf{A}^{-1} & \text{Inverse of a matrix} \\
\langle \mathbf{x}, \mathbf{y} \rangle & \text{Inner product of } \mathbf{x} \text{ and } \mathbf{y} \\
\mathbf{x} \cdot \mathbf{y} & \text{Dot product of } \mathbf{x} \text{ and } \mathbf{y} \\
\mathbf{B} = (b_1, b_2, b_3) & \text{(Ordered) tuple} \\
\mathbf{B} = [b_1, b_2, b_3] & \text{Matrix of column vectors stacked horizontally} \\
\mathbf{B} = \{b_1, b_2, b_3\} & \text{Set of vectors (unordered)} \\
\mathbb{Z}, \mathbb{N} & \text{Integers and natural numbers, respectively} \\
\mathbb{R}, \mathbb{C} & \text{Real and complex numbers, respectively} \\
\mathbb{R}^n & \text{\(n\)-dimensional vector space of real numbers} \\
\forall x & \text{Universal quantifier: for all } x \\
\exists x & \text{Existential quantifier: there exists } x \\
a := b & a \text{ is defined as } b \\
a =: b & b \text{ is defined as } a \\
a \propto b & a \text{ is proportional to } b, \text{ i.e., } a = \text{constant} \cdot b \\
g \circ f & \text{Function composition: ``\(g\) after \(f\)''} \\
\Leftrightarrow & \text{If and only if} \\
\Rightarrow & \text{Implies} \\
\mathcal{A}, \mathcal{C} & \text{Sets} \\
a \in \mathcal{A} & a \text{ is an element of set } \mathcal{A} \\
\emptyset & \text{Empty set} \\
\mathcal{A} \setminus \mathcal{B} & \mathcal{A} \text{ without } \mathcal{B} \\
D & \text{Number of dimensions; indexed by } d = 1, \ldots, D \\
N & \text{Number of data points; indexed by } n = 1, \ldots, N \\
\mathbf{I}_m & \text{Identity matrix of size } m \times m \\
\mathbf{0}_{m,n} & \text{Matrix of zeros of size } m \times n \\
\mathbf{1}_{m,n} & \text{Matrix of ones of size } m \times n \\
\mathbf{e}_i & \text{Standard/canonical vector (where \(i\) is the component that is 1)} \\
\dim & \text{Dimensionality of vector space} \\
\text{rk}(\mathbf{A}) & \text{Rank of matrix } \mathbf{A} \\
\text{Im}(\boldsymbol{\Phi}) & \text{Image of linear mapping } \boldsymbol{\Phi} \\
\ker(\boldsymbol{\Psi}) & \text{Kernel (null space) of a linear mapping } \boldsymbol{\Psi} \\
\text{span}[b_i] & \text{Span (generating set) of } b_i \\
\text{tr}(\mathbf{A}) & \text{Trace of } \mathbf{A} \\
\det(\mathbf{A}) & \text{Determinant of } \mathbf{A} \\
|\cdot| & \text{Absolute value or determinant (depending on context)} \\
\|\cdot\| & \text{Norm; Euclidean, unless specified} \\
\lambda & \text{Eigenvalue or Lagrange multiplier} \\
E_\lambda & \text{Eigenspace corresponding to eigenvalue } \lambda \\
\mathbf{x} \perp \mathbf{y} & \text{Vectors } \mathbf{x} \text{ and } \mathbf{y} \text{ are orthogonal} \\
V & \text{Vector space} \\
V^\perp & \text{Orthogonal complement of vector space } V \\
\sum_{n=1}^N x_n & \text{Sum of the } x_n: x_1 + \ldots + x_N \\
\prod_{n=1}^N x_n & \text{Product of the } x_n: x_1 \cdot \ldots \cdot x_N \\
\boldsymbol{\theta} & \text{Parameter vector} \\
\frac{\partial f}{\partial x} & \text{Partial derivative of } f \text{ with respect to } x \\
\frac{df}{dx} & \text{Total derivative of } f \text{ with respect to } x \\
\nabla & \text{Gradient} \\
f_* = \min_x f(x) & \text{The smallest function value of } f \\
x_* \in \arg\min_x f(x) & \text{The value } x_* \text{ that minimizes } f \text{ (note: arg min returns a set of values)} \\
\mathcal{L} & \text{Lagrangian} \\
\ell & \text{Negative log-likelihood} \\
\binom{n}{k} & \text{Binomial coefficient, } n \text{ choose } k \\
\text{Var}_X[\mathbf{x}] & \text{Variance of } \mathbf{x} \text{ with respect to the random variable } X \\
\mathbb{E}_X[\mathbf{x}] & \text{Expectation of } \mathbf{x} \text{ with respect to the random variable } X \\
\text{Cov}_{X,Y}[\mathbf{x}, \mathbf{y}] & \text{Covariance between } \mathbf{x} \text{ and } \mathbf{y} \\
X \perp Y \mid Z & X \text{ is conditionally independent of } Y \text{ given } Z \\
X \sim p & \text{Random variable } X \text{ is distributed according to } p \\
\mathcal{N}(\boldsymbol{\mu}, \boldsymbol{\Sigma}) & \text{Gaussian distribution with mean } \boldsymbol{\mu} \text{ and covariance } \boldsymbol{\Sigma} \\
\text{Ber}(\mu) & \text{Bernoulli distribution with parameter } \mu \\
\text{Bin}(N, \mu) & \text{Binomial distribution with parameters } N, \mu \\
\text{Beta}(\alpha, \beta) & \text{Beta distribution with parameters } \alpha, \beta \\
\hline
\end{array}
$$
